### cleaning price outliers

```python
# Box Plot
(df['price']).plot(kind="box", subplots=True, layout=(19,19), figsize=(80,80))
```

```python
# IQR based filtering
Q1 =df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Finding upper and lower limit
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

#Finding Outliers
df[df['price'] > upper_bound]
df[df['price'] < lower_bound]

#Trimming
new_df = df[df['price'] < upper_bound]
new_df.shape
```

```python
#Compare the plots after trimming

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['price'])
plt.subplot(2,2,2)
sns.boxplot(df['price'])
plt.subplot(2,2,3)
sns.distplot(new_df['price'])
plt.subplot(2,2,4)
sns.boxplot(new_df['price'])
plt.show()
```

```python
#Capping(Winsorization)

new_df_cap = df.copy()
new_df_cap['price'] = np.where(
    new_df_cap['price'] > upper_bound,
    upper_bound,
    np.where(
        new_df_cap['price'] < lower_bound,
        lower_bound,
        new_df_cap['price']
    )
)
#Compare the plots after capping

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['price'])
plt.subplot(2,2,2)
sns.boxplot(df['price'])
plt.subplot(2,2,3)
sns.distplot(new_df_cap['price'])
plt.subplot(2,2,4)
sns.boxplot(new_df_cap['price'])
plt.show()
```

```python
# Principal Component Analysis

from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
#Scale the data
scaled_data = scaler.transform(df)
```

```python
pca = PCA(n_components=4)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```
